### Clean up dataset with number string

```python
#change the variable type to float including taking out $ and number string with comma

tmdb_budget["production_budget"] = tmdb_budget["production_budget"].str.replace("$", "")
tmdb_budget["production_budget"] = tmdb_budget["production_budget"].str.replace(",", "")
tmdb_budget["production_budget"] = tmdb_budget["production_budget"].astype(float)
tmdb_budget["domestic_gross"] = tmdb_budget["domestic_gross"].str.replace("$", "")
tmdb_budget["domestic_gross"] = tmdb_budget["domestic_gross"].str.replace(",", "")
tmdb_budget["domestic_gross"] = tmdb_budget["domestic_gross"].astype(float)
tmdb_budget["worldwide_gross"] = tmdb_budget["worldwide_gross"].str.replace("$", "")
tmdb_budget["worldwide_gross"] = tmdb_budget["worldwide_gross"].str.replace(",", "")
tmdb_budget["worldwide_gross"] = tmdb_budget["worldwide_gross"].astype(float)
```
